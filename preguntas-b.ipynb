```python
import pandas as pd
pd.options.display.max_columns = None
import re
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import sidetable

plt.rcParams["figure.figsize"] = (10,10)
sns.set_palette("husl", 9)
```

```python
df_prof = pd.read_csv("../Archivos/profesionales_nonull.csv", index_col= 0)
```

```python
df_noprof = pd.read_csv("../Archivos/no_profesionales.csv", index_col= 0)
```

```python
columnas_b = df_noprof.columns[-7:].tolist()
```

```python
columnas_b
```

```python
def data_pregunta(dataf, colum):
    df = dataf[["genero", "puesto_trabajo", colum]]
    df[colum] = df[colum].str.split(",")
    df = df.explode(colum)
    return df
```

```python
df_q29b_noprof = data_pregunta(df_noprof, 'q29b_productos_familiar_cloud_computing')
#df_q31b_noprof = data_pregunta(df_noprof, 'q31b_productos_familiar_ml')
df_q32b_noprof = data_pregunta(df_noprof, 'q32b_productos_familiar_bd')
df_q34b_noprof = data_pregunta(df_noprof, 'q34b_herramientas_familiar_bi')
#df_q36b_noprof = data_pregunta(df_noprof, 'q36b_cat_herramientas_familiar_ml')
df_q37b_noprof = data_pregunta(df_noprof, 'q37b_cat_herramientas_familiar_ml_aut')
#df_q38b_noprof = data_pregunta(df_noprof, 'q38b_herramientas_familiar_gest_exp_ml')
```

```python
df_q29b_prof = data_pregunta(df_prof, 'q29b_productos_familiar_cloud_computing')
#df_q31b_prof = data_pregunta(df_noprof, 'q31b_productos_familiar_ml')
df_q32b_prof = data_pregunta(df_prof, 'q32b_productos_familiar_bd')
df_q34b_prof = data_pregunta(df_prof, 'q34b_herramientas_familiar_bi')
#df_q36b_prof = data_pregunta(df_noprof, 'q36b_cat_herramientas_familiar_ml')
df_q37b_prof = data_pregunta(df_prof, 'q37b_cat_herramientas_familiar_ml_aut')
#df_q38b_prof = data_pregunta(df_noprof, 'q38b_herramientas_familiar_gest_exp_ml')
```

# PRODUCTOS DE CLOUD COMPUTING
---

```python
orden_q29b = df_q29b_noprof['q29b_productos_familiar_cloud_computing'].value_counts().index.tolist()
```

```python
fig, axes = plt.subplots(2,1, sharex= True)

sns.countplot(data = df_q29b_noprof, y = 'q29b_productos_familiar_cloud_computing', ax = axes[0], order= orden_q29b)
sns.countplot(data = df_q29b_prof, y = 'q29b_productos_familiar_cloud_computing', ax = axes[1], order= orden_q29b)

fig.suptitle("PRODUCTOS DE CLOUD COMPUTING")
axes[0].set_title("No profesionales")
axes[1].set_title("Profesionales")

axes[0].set(xlabel=None)
axes[0].set(ylabel=None)
axes[1].set(xlabel=None)
axes[1].set(ylabel=None)

plt.tight_layout()
plt.show();
```
